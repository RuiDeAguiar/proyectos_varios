```python
#Creo la funcion que envia el correo, #, attachments=[]
def send_mail(sender, password, recipient, subject, body):
    msg = MIMEMultipart()
    msg['From'] = sender
    msg['To'] = recipient
    msg['Subject'] = subject

    msg.attach(MIMEText(body, 'plain'))

    # Conexi√≥n y Env√≠o SMTP
    try:
        print("üåç Intentando conectar con el servidor SMTP de Gmail...")
        
        # Conexi√≥n al servidor SMTP de Gmail (Host: 587, Protocolo: STARTTLS)
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls() # Iniciar el cifrado TLS
        
        # Iniciar sesi√≥n
        server.login(sender, password)
        print("‚úÖ Inicio de sesi√≥n exitoso.")
        
        # Enviar el correo
        text = msg.as_string() # Convertir el objeto msg a cadena de texto
        server.sendmail(sender, recipient, text)
        print(f"üìß Correo de texto enviado exitosamente a {recipient}!")
        
    except smtplib.SMTPAuthenticationError:
        print("‚ùå Error de Autenticaci√≥n SMTP. Revisa tu Contrase√±a de Aplicaci√≥n de 16 caracteres.")
    except Exception as e:
        print(f"‚ùå Ocurri√≥ un error al enviar el correo: {e}")
    finally:
        # Cerrar la conexi√≥n
        if 'server' in locals():
            server.quit()
            print("üëã Conexi√≥n SMTP cerrada.")
```
